# El algoritmo del lechero

Dispone de un camión con un cierto límite de peso, y un grupo de vacas disponibles para la venta. Cada vaca puede tener un peso distinto, y producir una cantidad diferente de leche al día.

Debes elegir qué vacas comprar y llevar en su camión, de modo que pueda **maximizar la producción de leche**, observando el límite de peso del camión.

1.- Para este propósito tienes que definir las siguientes **entradas**:

Entrada: Número total de vacas en la zona de Tolosa que están a la venta.

Entrada: Peso total que el camión puede llevar.

Entrada: Lista de pesos de las vacas.

Entrada: Lista de la producción de leche por vaca, en litros por día.

2.- El algoritmo que programes tiene que calcular la siguiente **salida**:

Salida: Cantidad máxima de producción de leche se puede obtener.


```
import numpy as np

class ProduccionLeche():
  def __init__(self, n_vacas, max_peso_camion, arr_pesos_vacas, arr_leche_dia ):
    '''
    Inicializamos las variables que vamos a necesitar.
    '''
    
    self.n_vacas = n_vacas
    self.max_peso_camion = max_peso_camion
    self.arr_pesos_vacas = np.array(arr_pesos_vacas)
    self.arr_leche_dia = np.array(arr_leche_dia)
    self.peso_vacas = 0
    self.total_leche = 0
    self.total_leche_max = 0

    
  def combinacion(self):
    '''
    Obtenemos todas las posibles combinaciones que se pueden dar al sumar elementos del vector arr_pesos_vacas.
    Para estar seguros de que obtenemos la máxima producción de leche, tenemos que pasar por todas combinaciones, a pesar de que pueda ser un proceso un tanto largo.
    
    Referencia: https://stackoverrun.com/es/q/68446

    '''
    import itertools
    vectorComb=[]
    for L in range(0, len(self.arr_pesos_vacas)+1):
     for subset in itertools.combinations( self.arr_pesos_vacas, L):
      vectorComb.append(subset)
       
    return vectorComb



  def index_kg_en_leche_array(self, value, array):
    '''
    Devuelve los índices en los que se encuentra un determinado value en el array.
    '''
    return np.where(np.in1d(array, value))[0]



  def maximizar_produccion(self, vectorComb):
    '''
    El objetivo del problema es maximizar los litros de leche. Pero primero tenemos que comprobar que la suma del peso de las vacas sea menor que el maximo que admite el camion.
    '''
    self.total_leche_max = 0
    for i in range (0, len(vectorComb)):
      suma_kg = 0
      for j in range (0, len(vectorComb[i])):

        suma_kg += vectorComb[i][j]
      
      '''
      Cuando se cumple la condición de que el peso de las vacas sea menor, calculamos los litros de producción de leche que podemos tener con esas vacas. Para eso, obtenemos los indices de las vacas escogidas en el array, y nos quedamos con los litros que 
      nos dan esos indices en el array de la leche(corresponden a las vacas escogidas)

      '''
      
      if suma_kg <= self.max_peso_camion:
        self.total_leche = 0
        for j in range(0, len(vectorComb[i])):

          index = []
          index = self.index_kg_en_leche_array(vectorComb[i][j], self.arr_pesos_vacas)

          '''
          *** PROBLEMA: cuando se repite un mismo número, al hallar el índice de ese valor en el array, obtenemos todas las posiciones en las cual aparece el valor en el array. Como estoy trabajando con un índice por cada ciclo, en las siguientes líneas he escrito un parche 
          que posiblemente corrija esto, aunque al no ser esta parte muy robusta, puede dar lugar a error.
          De hecho, solo consideramos el caso en el que se repitan 2, si son más daría error.
          Si ejecutas varias veces estas líneas de comando, podrás ver como en el 2nd y 3er ejemplo, los litros de leche cambian un poco, debido a este problema. ***
          '''
          if len(index)>1:
            #coger uno de los 2 al 50%
            from random import random
            if random() > 0.5:
              index = index[0]
            else:
              index = index[1]
        
          self.total_leche += self.arr_leche_dia[index]
     
        if self.total_leche > self.total_leche_max:
          self.total_leche_max = self.total_leche
        
    print("Max. Leche litros: ", self.total_leche_max)
        

def main():
  
  print('RESULTADO')
  print('===========================================================================')
  print('1er EJEMPLO')
  object1 = ProduccionLeche(6, 700, [360,250,400,180,50,90], [40,35,43,28,12,13])
  vectorComb = object1.combinacion()
  object1.maximizar_produccion(vectorComb)

  print('---------------------------------------------------------------------------')
  print('2do EJEMPLO')
  object2 = ProduccionLeche(8, 1000, [223,243,100,200,200,155,300,150], [30,34,28,45,31,50,29,1])
  vectorComb = object2.combinacion()
  object2.maximizar_produccion(vectorComb)

  print('---------------------------------------------------------------------------')
  print('3er EJEMPLO')
  object3 = ProduccionLeche(10, 2000, [340,355,223,243,130,240,260,155,302,130], [45,50,34,39,29,40,30,52,31,1])
  vectorComb = object3.combinacion()
  object3.maximizar_produccion(vectorComb)

  print('===========================================================================')
  print('NOTA: el resultado del 2do y 3er ejemplo puede no ser exacto, debido a la problemática explicada anteriormente.')
  print('A mejorar: solución robusta cuando se repiten números. En el ejemplo 2 --> se repite el nº 200. En el 3er ejemplo, el nº 130.')

main()
```
